### Xi_treecorr

compute correlation function $\xi(r)$, given halo and snapshot particle information

```python
### DMO case
halo_cat_dmo = Cat.matched_halo_cat_dmo
snap_cat_dmo = snap_dmo.snap_cat
Lbox = snap_dmo.Lbox

Xi_dmo = Xi_treecorr(halo_cat_dmo, snap_cat_dmo, Lbox)
```

```python
r_dmo, xi_dmo, varxi_dmo = Xi_dmo.run_treecorr()
```

```python
### hydro case
halo_cat_hydo = Cat.matched_halo_cat_hydro
snap_cat_hydro = snap_hydro.snap_cat
Lbox = snap_hydro.Lbox

Xi_hydro = Xi_treecorr(halo_cat_hydo, snap_cat_hydro, Lbox)
```

```python
%time r_hydro, xi_hydro, varxi_hydro = Xi_hydro.run_treecorr()
```

```python
%pylab inline

fig = plt.figure(figsize=(7.2,6.5))
plt.rc('font',size=20)

ax1=fig.add_axes([0.16,0.11,0.82,0.84])

ax1.plot(r_hydro, xi_hydro,label='hydro halo:[14,15]',color='b')
ax1.plot(r_dmo, xi_dmo,label='dmo matched',color='k')

ax1.set_xlabel('r [kpc/h]')
ax1.set_ylabel(r'$\xi(r)$')

ax1.set_yscale('log')
ax1.set_xscale('log')

ax1.legend(loc='best',prop={'size':18})
```

### $\xi(r) = \frac{\rho(r)-\bar\rho(r)}{\bar\rho(r)}$

### $\rho(r) = \bar\rho(r) (1+\xi(r))$

## $\frac{\rho_{\rm hydro}(r)}{\rho_{\rm dmo}(r)} = \frac{1+\xi_{\rm hydro}(r)}{1+\xi_{\rm dmo}(r)} \approx \frac{\xi_{\rm hydro}(r)}{\xi_{\rm dmo}(r)}$ 
 at small scale

```python
fig = plt.figure(figsize=(7.2,6.5))
plt.rc('font',size=20)

ax1=fig.add_axes([0.16,0.11,0.82,0.84])

ax1.plot(r_hydro, xi_hydro/xi_dmo,label='halo:[14,15]',color='b')
ax1.axhline(y=1,color='lightgray',zorder=-1)

ax1.set_xlabel('r [kpc/h]')
ax1.set_ylabel(r'$\xi_{hydro}(r)/\xi_{dmo}(r)$')

#ax1.set_yscale('log')
ax1.set_xscale('log')
ax1.set_title('TNG')

ax1.legend(loc='best',prop={'size':18})
```

```python
fig = plt.figure(figsize=(7.2,6.5))
plt.rc('font',size=20)

ax1=fig.add_axes([0.16,0.11,0.82,0.84])

ax1.plot(r_hydro, (xi_hydro+1)/(xi_dmo+1),label='halo:[14,15]',color='b')
ax1.axhline(y=1,color='lightgray',zorder=-1)

ax1.set_xlabel('r [kpc/h]')
ax1.set_ylabel(r'$\rho_{hydro}(r)/\rho_{dmo}(r)$')

#ax1.set_yscale('log')
ax1.set_xscale('log')
ax1.set_title('TNG')

ax1.legend(loc='best',prop={'size':18})
```
